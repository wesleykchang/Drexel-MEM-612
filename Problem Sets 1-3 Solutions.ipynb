1. Unsteady mass conservation equation: $D\rho/Dt\ +\ \rho \nabla  \overrightarrow{v}\ =\ 0$

$d\rho/dt\ +\ u d\rho/dx\ +\ vd\rho/dy\ +\ \rho (du/dx\ +\ dv/dy)$

the question asks what relationships must hold for the simplified mass continuity equation $du/dx\ +\ dv/dy\ =\ 0$ to hold, if the largest velocity gradient is du/dx and the largest density gradient is $d\rho/dx$ (such as in a case where laminar flow proceeds along only one axis and the other directions are negligible?)

we consider those two terms: $u d\rho/dx$, $\rho du/dx$

since the simplified continuity equation does not have a density dependence, we reason that: $\rho du/dx$ is much greater than $u d\rho/dx$

hence, $\rho / u\$ is greater than $d\rho/dx / du/dx$


this is another way of saying that the density gradients are relatively small


2. See plot below. Air has a thicker thermal boundary layer relative to its velocity (hydrodynamic) boundary layer, because it has a relatively low Pr (thermal diffusivity dominates). Water has a thinner boundary layer overall: it is more dense and has a higher Pr




```python
import matplotlib.pyplot as plt
import numpy as np


fig1 = plt.figure()
ax1 = fig1.add_subplot(111)

Pr_H2O = 4.6
Pr_air = 0.709

x = np.linspace(0,1)
print(x)

'''water'''
ax1.plot(x, 0.005*np.sqrt(x), c = 'k', label='water')
ax1.plot(x, 0.005*np.sqrt(x)* (Pr_H2O)**(-1/3), c='r', label='water, thermal b.l.')

'''air'''
ax1.plot(x, 0.005*np.sqrt(x)*np.sqrt(1.659E-5/6.6E-7), c='orange', label='air')
ax1.plot(x, 0.005*np.sqrt(x)*np.sqrt(1.659E-5/6.6E-7) *Pr_air**(-1/3), c='g', label='air, thermal b.l.')
ax1.legend()
plt.show()
```

3. The two boundary conditions are no slip at the inner and outer walls: $u\ =\ 0, r\ =\ r_i$, $u\ =\ 0,\ r\ =\ r_o$ where $r_i$ is the inner radius and $r_o$ is the outer wall radius. 

b) The general equation is: $u\ =\ r^2/4\mu (dP/dx)\ +\ c_1 ln(r)\ +\ c_2$

c) plugging in boundary conditions and solving for the two constants

$c_1\ ln(r_o/r_i)\ =\ -1/4\mu dP/dx(r_o^2\ -\ r_i^2)$

$c_2\ =\ -(r_o)^2/4\mu dP/dx\ -\ c_1 ln(r_o)$

d) shear stresses: $\tau _i\ =\ \mu (du/dr)_{r=r_i}$

$\tau _o\ =\ -\mu (du/dr)_{r=r_o}$

if the inner tube shrinks, the difference in the shear stresses should increase. plotting (see below), we see that a decreasing inner radius increases the shear stress of both walls (which is directed in the reverse direction of flow), with the shear stresss of the inner wall increasing more. Thus, the difference in shear stress will increase. 


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

r_i = np.linspace(1,10,10)

tau_i = r_i/2*1 - 0.25/r_i * (20**2 - r_i**2)/np.log(20/r_i)

tau_o = -20/2 * 1 + 1/80 * (20**2 - r_i**2)/np.log(20/r_i)

plt.plot(r_i, tau_i, c = 'b')

plt.plot(r_i, tau_o, c = 'g')

plt.show()
```
